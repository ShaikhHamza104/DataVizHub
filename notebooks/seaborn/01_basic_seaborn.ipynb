```python
import  pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import  plotly.express as px
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
# scatter plot -> axes level function
sns.scatterplot(tips,x = 'total_bill', y = 'tip',hue='sex',size='size')
```

```python
# scatter plot -> axes level function
sns.relplot(data=tips,x = 'total_bill', y = 'tip',kind = 'scatter',hue='sex',style='time',size='size')
```

```python
# line plot
gap = px.data.gapminder()
india = gap[gap['country']=='India']
sns.lineplot(india, x ='year',y = 'lifeExp')
```

```python
sns.relplot(data= india , x = 'year',y = 'lifeExp',kind='line')
```

```python
temp_df = gap[gap['country'].isin(['India','China','Pakistan'])]
```

```python
sns.relplot(data = temp_df, kind='line',x = 'year',y = 'lifeExp',hue='country')
```

```python
sns.lineplot(data = temp_df,x = 'year',y = 'lifeExp',hue='country')
```

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df.head()
```

```python
sns.relplot(kind='line', data= temp_df , x = 'year', y = 'lifeExp' ,hue='country', style='continent',size='continent')
```

```python
# facet plot
sns.relplot(kind='line',data=tips,x='total_bill',y = 'tip',col='sex',row = 'day')
```

```python
# col warp
sns.relplot(data = gap ,x = 'lifeExp',y = 'gdpPercap',kind='scatter',col='year',col_wrap=2)
```

```python
# figure level ->displot
# axes level -> hisplot -> kdeplot -> rugplot
sns.histplot(data=tips, x = 'total_bill')
```

```python
sns.displot(data=tips,kind='hist',x='total_bill')
```

```python
# bin size
sns.histplot(data=tips,x='total_bill',bins=20)
```

```python
# countplot
sns.histplot(data=tips,x='day')
```

```python
# hue
sns.histplot(data=tips,x='tip',hue='sex')
```

```python
# element -> step
sns.displot(data = tips,kind='hist',x = 'tip',element = 'step' )
```

```python
temp_df = gap.pivot_table(index = 'country',columns = 'year',values = 'lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
temp_df = gap[gap['continent']=='Europe'].pivot_table(index = 'country',columns = 'year',values = 'lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidths=0.5,cmap='summer')
```
