```python
from itertools import count

import  seaborn as sns
import  plotly.express as px
from seaborn import boxplot
import  numpy as np
from select import error

sns.set_theme('paper')
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
# strip plot -> A
sns.stripplot(data=tips, x='day', y='total_bill',jitter=True)
```

```python
# using catplot -> F
sns.catplot(data=tips, x='day', y='total_bill',kind='strip')
```

```python
sns.catplot(data=tips, x='day', y='total_bill',kind='strip',jitter=1.10,hue='sex')
```

```python
sns.swarmplot(data=tips, x = 'day', y = 'total_bill',hue = 'sex')
```

```python
# swarmplot
sns.catplot(data=tips, x='day', y='total_bill',kind='swarm',hue='sex')
```

```python
# boxplot
sns.boxenplot(data = tips,x = 'total_bill')
```

```python
# boxplot
sns.catplot(data = tips,x = 'day',y = 'total_bill',kind = 'box',hue='sex')
```

```python
# boxplot
sns.catplot(data = tips,x = 'total_bill',kind = 'box',hue='sex')
```

```python
# violin plot
sns.violinplot(data=tips,x = 'day', y = 'total_bill',hue='sex')
```

```python
sns.catplot(data=tips,x = 'day', y = 'total_bill',hue='sex',kind='violin',split=True)
```

```python
sns.barplot(data=tips,x = 'sex', y = 'total_bill',errorbar=None,hue='smoker',estimator=np.median)
```

```python
# point plot
sns.pointplot(data=tips,x = 'sex',y = 'total_bill',errorbar=None,hue='smoker',estimator=np.median)
```

```python
# count plot
sns.countplot(data=tips,x = 'sex')
```

```python
sns.catplot(data=tips,x = 'day',hue='sex',kind='count')
```

```python
sns.catplot(data=tips,x = 'tip',y = 'smoker',col='sex',kind='box',row='time',errorbar=None)
```

```python
sns.regplot(data=tips,x = 'total_bill',y = 'tip')
```

```python
sns.lmplot(data=tips,x = 'total_bill',y = 'tip',hue='sex')
```

```python
sns.residplot(data=tips, x = 'total_bill',y = 'tip')
```

```python
g =  sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.boxplot,'sex','total_bill')
g.add_legend()
g.fig.suptitle('Boxplot')
```

```python
sns.pairplot(data= iris,hue='species')
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
sns.jointplot(data=tips,x = 'total_bill',y = 'tip',hue = 'smoker')
```

```python
g = sns.JointGrid(data=tips,x = 'total_bill',y = 'tip',hue = 'smoker')
g.plot(sns.scatterplot,sns.histplot)
```
