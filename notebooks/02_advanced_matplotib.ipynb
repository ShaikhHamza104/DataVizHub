### Anotation

```python
batter = pd.read_csv('../datasets/batter.csv')
```

```python
batter.head()
```

```python
sample_df = batter.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.scatter(x,y,c='red',s=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.text(3,8,'Text',fontsize=12,c = 'blue')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s = sample_df['runs'])
plt.xlabel('Average')
plt.ylabel('Strike Rate')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i],color='red')
plt.show()
```

### Horizontal and Vertical Line

```python
plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'],sample_df['strike_rate'],sample_df['runs'])
plt.xlabel('Average')
plt.ylabel('Strike Rate')

plt.axhline(130, color='r', linestyle='--')
plt.axvline(30, color='r')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i],color='red')
plt.show()
```

### Subplots

```python
# a different way to plot graphs
fig, ax = plt.subplots(figsize=(18,10))
ax.scatter(sample_df['avg'],sample_df['strike_rate'],sample_df['runs'],color = 'yellow')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
fig.suptitle('Batter Performance')
plt.show()
```

### 3D Scatter Plot

```python
fig = plt.figure(figsize=(18,10))
ax = plt.subplot(projection='3d')
ax.scatter3D(batter['runs'], batter['avg'], batter['strike_rate'],color='green',s=10)
ax.set_title('Batter Performance')
ax.set_xlabel('Runs')
ax.set_ylabel('Average')
ax.set_zlabel('Strike Rate')
plt.show()
```

### 3D Line Plot

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
z = [11,12,13,14,15]
fig = plt.figure(figsize=(18,10))
ax = fig.add_subplot(111, projection='3d')
ax.plot3D(x, y, z, c='r', marker='o')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
```

### 3D surface plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(18,10))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

### Contour Plot

```python
fig = plt.figure(figsize=(18,10))
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

### Contourf Plot

```python
fig = plt.figure(figsize=(18,10))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```
